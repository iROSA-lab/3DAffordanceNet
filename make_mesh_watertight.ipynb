```python
import trimesh
import open3d as o3d
from open3d.web_visualizer import draw
import point_cloud_utils as pcu

mesh_path = 'data_root/3DGraspAff/obj_meshes/train/ShapeNet/9c930a8a3411f069e7f67f334aa9295c/models/model_normalized.obj'

# v is a nv by 3 NumPy array of vertices
# f is an nf by 3 NumPy array of face indexes into v
# load mesh with trimesh
v, f = pcu.load_mesh_vf(mesh_path)
len(f)
```

```python
# Optional resolution parameter (default is 20_000).
# See https://github.com/hjwdzh/Manifold for details
resolution = 20_000
v_watertight, f_watertight = pcu.make_mesh_watertight(v, f, resolution=resolution)
```

```python
len(f_watertight)
```

```python
# Save the watertight mesh
# mesh = trimesh.Trimesh(vertices=v_watertight, faces=f_watertight)
mesh = o3d.geometry.TriangleMesh()
mesh.vertices = o3d.utility.Vector3dVector(v_watertight)
mesh.triangles = o3d.utility.Vector3iVector(f_watertight)
# draw(mesh)
o3d.io.write_triangle_mesh('test.obj', mesh)
```

```python
# load same mesh with trimesh
mesh = trimesh.Trimesh(vertices=v_watertight, faces=f_watertight)
mesh.show()
```

```python
target_num_faces = f_watertight.shape[0] // 10  # Downsample by a factor of 10

# v_decimate, f_decimate are the vertices/faces of the decimated mesh
# v_correspondence, f_correspondence are the vertices and faces in the dense mesh which generated each
# downsampled vertex/face
v_decimate, f_decimate, v_correspondence, f_correspondence = pcu.decimate_triangle_mesh(v_watertight, f_watertight, target_num_faces)
mesh_decimated = trimesh.Trimesh(vertices=v_decimate, faces=f_decimate)
print(len(v_decimate), len(f_decimate))
mesh_decimated.show()
```
